### Data Preparation

##### removing Duplicates

```python
duplicate=df.duplicated()
duplicate.sum()
```

```python
df=df.drop(columns=["PassengerId","Name","Ticket","Cabin"],axis=1)
df.head()
```

```python
sns.boxplot(df["Age"])
```

```python
def remove_outlier(col):
    sorted(col)
    q1,q3=col.quantile([0.25,0.75])
    iqr=q3-q1
    lower_range=q1-(1.5*iqr)
    upper_range=q3+(1.5*iqr)
    return lower_range,upper_range
```

```python
lowage,highage=remove_outlier(df["Age"])
```

```python

df["Age"]=np.where(df["Age"]>highage,highage,df["Age"])
df["Age"]=np.where(df["Age"]<lowage,lowage,df["Age"])
```

```python
sns.boxplot(df["Age"])
```

```python
df.shape
```

```python
df.info()
```

```python
##### Missing values
median=df["Age"].median()
median
```

```python
df["Age"]=df["Age"].replace(np.nan,median)
```

```python
df.isnull().sum()
```

```python
mode=df["Embarked"].mode()[0]
mode
```

```python
df["Embarked"]=df["Embarked"].replace(np.nan,mode)
```

```python
df.isnull().sum()
```

##### Encoding

```python
df = pd.get_dummies(df, columns = ['Sex', 'Embarked', 'Pclass'])
```
