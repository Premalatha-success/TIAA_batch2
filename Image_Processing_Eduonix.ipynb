```python
import cv2
import numpy as np
import tkinter as tk
from tkinter import filedialog
from PIL import Image, ImageTk
```

```python
def apply_filter(filter_func):
    global current_image

    if current_image is None:
        return

    filtered_img = filter_func(current_image)
    update_display(filtered_img)
```

```python
def update_display(image):
    pil_image = Image.fromarray(image)
    pil_image = pil_image.resize((300, 300))
    tk_image = ImageTk.PhotoImage(pil_image)
    sketch_image_label.config(image=tk_image)
    sketch_image_label.image = tk_image

def open_image():
    global current_image
    file_path = filedialog.askopenfilename()
    if file_path:
        current_image = cv2.imread(file_path)
        update_display(current_image)
```

```python

def apply_cartoonize(img):
    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # Apply median blur to reduce noise
    gray = cv2.medianBlur(gray, 5)
    # Detect edges using adaptive thresholding
    edges = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9)
    # Combine the edges with the original image to create the cartoon effect
    cartoon = cv2.bitwise_and(img, img, mask=edges)
    return cartoon

def apply_vintage(img):
    # Apply Sepia effect with reduced saturation
    sepia_filter = np.array([[0.272, 0.534, 0.131],
                             [0.349, 0.686, 0.168],
                             [0.393, 0.769, 0.189]])
    sepia_image = cv2.transform(img, sepia_filter)
    sepia_image[np.where(sepia_image > 255)] = 255  # Cap values to 255
    # Apply vignette effect to the sepia image
    rows, cols, _ = sepia_image.shape
    mask = np.zeros((rows, cols), dtype=np.uint8)
    cv2.circle(mask, (cols // 2, rows // 2), int(rows * 0.7), 255, -1)
    vignette = cv2.bitwise_and(sepia_image, sepia_image, mask=mask)
    return vignette

def apply_pop_art(img):
    # Apply posterization effect with increased color channels
    posterize_img = img.copy()
    levels = 6
    posterize_img = posterize_img // (256 // levels) * (256 // levels)
    # Split the image channels and increase contrast for each channel
    b, g, r = cv2.split(posterize_img)
    b = cv2.equalizeHist(b)
    g = cv2.equalizeHist(g)
    r = cv2.equalizeHist(r)
    pop_art_img = cv2.merge((b, g, r))
    return pop_art_img

def pencil_sketch(img):
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Invert the grayscale image
    inverted_image = 255 - gray_image

    # Apply Gaussian blur to the inverted image
    blurred_image = cv2.GaussianBlur(inverted_image, (21, 21), 0)

    # Blend the grayscale image and the blurred image
    pencil_sketch_image = cv2.divide(gray_image, 255 - blurred_image, scale=256.0)

    return pencil_sketch_image

def apply_mirror_effect(img):
    # Mirror the image horizontally
    mirrored_img = cv2.flip(img, 1)
    return mirrored_img

def start_realtime_camera():
    cap = cv2.VideoCapture(0)

    while True:
        ret, frame = cap.read()

        if not ret:
            break

        # Apply the selected filter
        if current_filter is not None:
            filtered_frame = current_filter(frame)
        else:
            filtered_frame = frame

        # Convert frame to RGB for Tkinter display
        rgb_frame = cv2.cvtColor(filtered_frame, cv2.COLOR_BGR2RGB)

        # Update the display
        pil_image = Image.fromarray(rgb_frame)
        pil_image = pil_image.resize((300, 300))
        tk_image = ImageTk.PhotoImage(pil_image)
        sketch_image_label.config(image=tk_image)
        sketch_image_label.image = tk_image

        root.update()

    cap.release()

def download_image():
    global current_image

    if current_image is None:
        return

    save_path = filedialog.asksaveasfilename(defaultextension=".jpg")
    if save_path:
        cv2.imwrite(save_path, cv2.cvtColor(current_image, cv2.COLOR_BGR2RGB))

def select_filter(filter_func):
    global current_filter
    current_filter = filter_func

# Create the Tkinter root window
root = tk.Tk()
root.title("Real-Time Image Filter Application")
root.geometry("1000x500")

# Create buttons frame
buttons_frame = tk.Frame(root)
buttons_frame.pack(side=tk.TOP, pady=10)

# Create buttons for each filter
filters = [
    ("Cartoonize", apply_cartoonize),
    ("Vintage", apply_vintage),
    ("Pop Art", apply_pop_art),
    ("pencil_sketch", pencil_sketch),
    ("Mirror Effect", apply_mirror_effect),
]

current_filter = None
for text, filter_func in filters:
    filter_button = tk.Button(buttons_frame, text=text, command=lambda f=filter_func: select_filter(f))
    filter_button.pack(side=tk.LEFT, padx=5)

# Create image display frame
image_frame = tk.Frame(root)
image_frame.pack(side=tk.TOP, pady=10)

# Create label for the camera feed
sketch_image_label = tk.Label(image_frame)
sketch_image_label.pack()

# Start real-time camera feed in a separate thread
import threading
camera_thread = threading.Thread(target=start_realtime_camera)
camera_thread.start()

# Create Save Image button
save_button = tk.Button(root, text="Save Image", command=download_image, padx=20)
save_button.pack(pady=10)

# Start the Tkinter main loop
current_image = None
root.mainloop()
```
