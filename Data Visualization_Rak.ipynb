### Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x=np.array([1,2])
y=np.array([5,10])

plt.plot(x,y)
plt.show()
```

```python
x=np.array([1,2])
y=np.array([5,10])

plt.plot(x,y,"o")
plt.show()
```

```python
x=np.array([1,2])
y=np.array([5,10])

plt.plot(x,y,'*''g')
plt.show()
```

```python
x=np.array([1,2])
y=np.array([5,10])

plt.plot(x,y,marker="*",ms=48,color="pink",mec="k",linestyle="dashed")
plt.show()
```

```python
y1=np.array([3,1,6,2])
y2=np.array([2,4,1,6])
plt.plot(y1,color="k",marker="*")
plt.plot(y2,color="red",marker="o")
plt.show()
```

```python
y1=np.array([3,1,6,2])
y2=np.array([2,4,1,6])
plt.plot(y1,color="k",marker="*")
plt.plot(y2,color="red",marker="o")
plt.title("Sample",fontsize=20,color="g")
plt.show()
```

```python
y1=np.array([3,1,6,2])
y2=np.array([2,4,1,6])
plt.plot(y1,color="k",marker="*")
plt.plot(y2,color="red",marker="o")
plt.title("Sample",fontsize=20,color="g")
plt.xlabel("X",loc="right")
plt.ylabel("Y",loc="top")
plt.show()
```

```python
y1=np.array([3,1,6,2])
y2=np.array([2,4,1,6])
plt.plot(y1,color="k",marker="*")
plt.plot(y2,color="red",marker="o")
plt.title("Sample",fontsize=20,color="g")
plt.xlabel("X",loc="right")
plt.ylabel("Y",loc="top")
plt.grid(color="g",linestyle="dotted")
plt.show()
```

```python
y1=np.array([3,1,6,2])
y2=np.array([2,4,1,6])
plt.plot(y1,color="k",marker="*",linewidth=6.5)
plt.plot(y2,color="red",marker="o")
plt.title("Sample",fontsize=20,color="g")
plt.xlabel("X",loc="right")
plt.ylabel("Y",loc="top")
plt.grid(color="g",linestyle="dotted",axis="y")
plt.show()
```

```python
X=np.array(["A","B","C"])
y=np.array([1,2,3])
plt.bar(X,y,color="k")
plt.show()
```

```python
X=np.array(["A","B","C"])
y=np.array([1,2,3])
plt.barh(X,y,color="k")
plt.show()
```

```python
y=np.array([10,23,15,21,20])
mylabels=["A","B","C","D","E"]
mycolors=["black","brown","yellow","aqua","blue"]
myexplode=[0,0.15,0.2,0.3,0.15]
plt.pie(y,startangle=90,explode=myexplode,colors=mycolors)
plt.legend(labels=mylabels)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
pd.set_option("display.max_rows",None)
pd.set_option("display.max_columns",None)
```

```python
import matplotlib
matplotlib.rcParams.update({'font.size':15})
plt.style.use("dark_background")
```

```python
df=pd.read_csv("https://gist.githubusercontent.com/ryanorsinger/0c702f67dfe59346cd4677cd24f7aec7/raw/429f0d47856670e57b31d1f78a1abc54c889c7ca/tips.csv")
```

```python
df.head()
```

```python
### Countplot-Sex
sns.countplot(x="sex",data=df,palette="Paired")
plt.show()
```

```python
sns.barplot(x="day",y="total_bill",data=df,palette="Dark2")
```

```python
sns.boxplot(x="day",y="total_bill",hue="sex",data=df)
plt.show()
```

```python
sns.kdeplot(x="total_bill",hue="sex",data=df)
```

```python
sns.stripplot(x="time",y="total_bill",hue="sex",data=df)
```

```python
sns.scatterplot(x="total_bill",y="tip",data=df,hue="sex")
```

```python
sns.swarmplot(x="day",y="total_bill",hue="sex",data=df)
plt.show()
```

```python
sns.lineplot(x="size",y="total_bill",hue="sex",data=df)
plt.show()
```

```python
sns.countplot(x="smoker",hue="sex",data=df)
```
